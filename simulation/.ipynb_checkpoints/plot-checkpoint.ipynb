```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {r'$\alpha$': 1, r'$\delta$': 3, r'$\gamma_w$': 0.45, r'$\gamma_b$': -0.5}
data_mat = np.array([[[ 0.9732,2.9796,0.4332,-0.5053],[0.96,2.973,0.426,-0.503]],
                    [[1.1844,3.0209,0.3968,-0.4558],[0.96,2.971,0.425,-0.504]]])
err_mat = np.array([[[0.005,0.0035,0.009,0.0092],[0.006,0.004,0.006,0.006]],
                   [[0.0112,0.0028,0.0163,0.021],[0.004,0.005,0.009,0.006]]])

names = list(data.keys())
values = np.array(list(data.values()))

fig, axs = plt.subplots(2, 2, figsize=(9,9))

for i in range(2):
    for j in range(2):
        axs[i][j].errorbar(data_mat[i][j]-values, names,fmt='bo',xerr=1.96*err_mat[i][j],ms=3.5)
        axs[i][j].set(xlabel=r'$\hat{\theta}_{MLE}-\theta_{Truth}$')
        axs[i][j].set_xlim([-0.21, 0.21])
        axs[i][j].axvline(0,c='r',ls='--',lw=1)

axs[0][0].set(ylabel='Homophily in model')
axs[0][0].set_title('Homophily in data')
axs[0][1].set_title('Homophily not in data')
axs[1][0].set(ylabel='Homophily not in model')


fig.suptitle('Model Inference with Homophily Effects')
plt.show()
```
